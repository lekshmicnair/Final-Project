```python

import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style="darkgrid")
```

# Reading and Cleaning Data

```python
# Load the Math and Portuguese data
math_df = pd.read_csv("data/student-mat.csv", sep=';')
port_df = pd.read_csv("data/student-por.csv", sep=';')
```

```python
port_df
```

```python
math_df
```

```python
# Combine the data
student_data_df = pd.concat([math_df,port_df])
student_data_df
```

```python
# Rename  the columns
student_data_df.columns = ['school','sex','age','address','family_size','parents_status','mother_education','father_education',
           'mother_job','father_job','reason','guardian','travel_time','study_time','failures','school_support',
          'family_support','paid_classes','extra_activities','nursery_school','wants_higher_edu','internet','romantic','family_relation',
          'free_time','go_out','weekday_alcohol','weekend_alcohol','health','absences','Q1_score','Q2_score','final_score']
```

```python
# Create final grade column for letter grades A+:18-20 A:16-17 B:14-15 C:10-13 & F<10 
student_data_df['final_letter_grade'] = 'None'
student_data_df.loc[(student_data_df.final_score >= 18) & (student_data_df.final_score <= 20), 'final_letter_grade'] = 'A+' 
student_data_df.loc[(student_data_df.final_score >= 16) & (student_data_df.final_score <= 17), 'final_letter_grade'] = 'A'
student_data_df.loc[(student_data_df.final_score >= 14) & (student_data_df.final_score <= 15), 'final_letter_grade'] = 'B' 
student_data_df.loc[(student_data_df.final_score >= 10) & (student_data_df.final_score <= 13), 'final_letter_grade'] = 'C' 
student_data_df.loc[(student_data_df.final_score >= 0) & (student_data_df.final_score <= 10), 'final_letter_grade'] = 'F' 
student_data_df
```

```python
#Check for null
student_data_df.isnull().values.any()
```

# Final Grade distribution and Correlation Plots

```python
# Final grade plot- seaborn count plot

plt.figure(figsize=(10,10))
sns.countplot(student_data_df.final_letter_grade,order=["A+","A","B","C","F"], palette='Set2')
plt.xlabel('Final Grade',fontsize=15)
plt.ylabel('Count of Students',fontsize=15)
plt.title('Final Grade Distribution',fontsize=20)

plt.savefig ('../Visualizations/Final_grade.png')
```

# Initial Model


```python
#Classification -Imports and Data Preprocessing

from sklearn import tree
from sklearn.preprocessing import LabelEncoder

Classification_df= student_data_df.drop(['final_score'], axis=1)
label = LabelEncoder()
Classification_df.final_letter_grade = label.fit_transform(Classification_df.final_letter_grade)
Classification_df
```

```python
#Assign X and y values
X = Classification_df.drop('final_letter_grade',axis=1)
y = Classification_df.final_letter_grade
print(X.shape, y.shape)

#train test split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=56)
len(list(X_train))
```

```python
#encode multiple columns using get_dummies
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
feature_names = X_train.columns
print(X_train.shape, X_test.shape)
```

```python
X_train.columns
```

```python
y_train
```

```python
#Random Forest Classifier Model and Fit
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
rf = RandomForestClassifier(n_estimators=500)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
#See the selected features
sel = SelectFromModel(RandomForestClassifier(n_estimators = 500))
sel.fit(X_train, y_train)
sel.get_support()

selected_feat= X_train.columns[(sel.get_support())]
len(selected_feat)

selected_feat
```

```python
#Sort the important features by weight
important_factors=sorted(zip(rf.feature_importances_, feature_names), reverse=True)
important_factors
```

# Analyse the main factors and Create Visualizations

```python

important_factors_full_df = pd.DataFrame (important_factors,columns=['Weight','Factor'])
important_factors_df= important_factors_full_df.head(10)

important_factors_df
```

```python
# Bar Plot for important features
plt.figure(figsize=(15,15))
plt.xticks(range(len(important_factors_df['Weight']*100)), important_factors_df['Factor'])
plt.xlabel('Factor',fontsize=15)
plt.ylabel('Weights',fontsize=15)
plt.title('Important Feature',fontsize=20)
plt.bar(range(len(important_factors_df['Weight']*100)), (important_factors_df['Weight']*100)) 
plt.show()
```

# Linear regression on Period 1, Period 2 and Final Score

```python
#Simple linear regression with Final score as the dependent variable and Period one as the independent variable

X_new = student_data_df.iloc[:,30:31].values
y_new = student_data_df.iloc[:,-2].values

#Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new,y_new,random_state = 42)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_new, y_train_new)

#Predicting the Test set results
predictions_new = model.predict(X_test_new)
score_new= model.score(X_test_new, y_test_new)
print(f"True output: {y_test_new[0]}")
print(f"Predicted output: {predictions_new[0]}")
print(f"Prediction Error: {predictions_new[0]-y_test_new[0]}")
print(f"Rsquare score: {score_new}")

predicted_df=pd.DataFrame({"Predicted": predictions_new, "Actual": y_test_new, "Error": predictions_new - y_test_new})[["Predicted", "Actual", "Error"]]
predicted_df
```

```python
#Visualizing the Training data results
plt.figure(figsize=(10,10))
plt.scatter(X_train_new, y_train_new, color = 'blue')
plt.plot(X_train_new,(model.predict(X_train_new)), color ='red')
plt.title('Final score vs Q1 Score (Training data)',fontsize=20)
plt.xlabel('Q1 Score',fontsize=15)
plt.ylabel('Final score',fontsize=15)
plt.show()
```

```python
#Visualizing the Testing data results
plt.figure(figsize=(10,10))
plt.scatter(X_test_new, y_test_new, color = 'blue')
plt.plot(X_train_new,(model.predict(X_train_new)), color ='red')
plt.title('Final score vs Q1 Score (Testing data)',fontsize=20)
plt.xlabel('Q1 Score',fontsize=15)
plt.ylabel('Final score',fontsize=15)

plt.savefig ('../Visualizations/Q1_final.png')
```

# Urban or Rural?

```python
# Address: urban vs. rural 
student_data_df.address = student_data_df.address.map({'U':'Urban', 'R':'Rural'})

plt.figure(figsize=(10,10))
sns.countplot(student_data_df.address, facecolor=(0, 0, 0, 0),linewidth=5,edgecolor=sns.color_palette("dark", 2))                 
plt.title('Students Count in Urban and Rural Area', fontsize=20)
plt.xlabel('Address Type', fontsize=15)
plt.ylabel('Number Of Students', fontsize=15)

plt.savefig ('../Visualizations/urban_rural.png')
```

```python
#Plot final grade by address area
index = ['A+','A','B','C','F']
student_address = pd.crosstab(index=student_data_df.final_letter_grade, columns=student_data_df.address)
student_address
percentage_student_address = student_address.apply(lambda col: col/col.sum()).reindex(index)
percentage_student_address.plot.bar(colormap="Paired", fontsize=15, figsize=(10,10))
plt.title('Final Grade By Address Area', fontsize=20)
plt.ylabel('Percentage of Students', fontsize=15)
plt.xlabel('Final Grade', fontsize=15)

plt.savefig ('../Visualizations/urban_rural_grades.png')
```
